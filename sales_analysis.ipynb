```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Order_Date": [
        "2024-01-15", "2024-01-18", "2024-02-05", "2024-02-20",
        "2024-03-10", "2024-03-18", "2024-04-02", "2024-04-12",
        "2024-05-05", "2024-05-22", "2024-06-14", "2024-06-25"
    ],
    "Customer": [
        "John Smith", "Mary Johnson", "David Lee", "John Smith",
        "Sarah Brown", "Michael Adams", "Emily Clark", "David Lee",
        "Daniel Wilson", "Olivia Taylor", "Mary Johnson", "James White"
    ],
    "Region": [
        "North", "South", "West", "North",
        "East", "North", "South", "West",
        "North", "East", "South", "West"
    ],
    "Product": [
        "Laptop", "Desk", "Printer", "Monitor",
        "Laptop", "Keyboard", "Desk Chair", "Mouse",
        "Laptop", "Headset", "Monitor", "Desk"
    ],
    "Quantity": [2,1,3,1,1,5,2,4,1,3,2,1],
    "Unit_Price": [900,400,250,300,950,45,180,25,920,70,310,420]
}

df = pd.DataFrame(data)
df.head()
```

```python
df["Revenue"] = df["Quantity"] * df["Unit_Price"]
df
```

```python
df["Revenue"].sum()
```

```python
df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Region")["Revenue"].sum().sort_values(ascending=False)
```

```python
region_revenue = df.groupby("Region")["Revenue"].sum()

region_revenue.plot(kind="bar")
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.show()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Month"] = df["Order_Date"].dt.to_period("M")

monthly_revenue = df.groupby("Month")["Revenue"].sum()

monthly_revenue.plot(kind="line")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```
